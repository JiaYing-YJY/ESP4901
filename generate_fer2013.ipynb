```python
import torch
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt
from random import randint
```

```python
trainset = torchvision.datasets.FER2013(root='./data', split="train",
                                       transform=transforms.ToTensor())
testset = torchvision.datasets.FER2013(root='./data', split="test",
                                       transform=transforms.ToTensor())
print(len(trainset))
print(len(testset))
```

```python
idx=4
pic, label =trainset[idx]
pic=pic.squeeze()
plt.imshow(pic.numpy(), cmap='gray')
plt.show()
print(label)
```

```python
train_data=torch.Tensor(28709,48,48)
train_label=torch.LongTensor(28709)
for idx , example in enumerate(trainset):
    train_data[idx]=example[0].squeeze()
    train_label[idx]=example[1]
torch.save(train_data,'data/total_data.pt')
torch.save(train_label,'data/data_label.pt')
```
